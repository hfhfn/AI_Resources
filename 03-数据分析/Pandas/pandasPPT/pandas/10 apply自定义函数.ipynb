## Series的apply方法

```python
# 数据准备
import pandas as pd
df = pd.DataFrame({'a':[10,20,30],'b':[20,30,40]})
```

```python
# 创建一个自定义函数
def my_sq(x):
    return x**2
```

```python
#apply方法有一个func参数, 把传入的函数应用于Series的每个元素
sq = df['a'].apply(my_sq)
#注意,把my_sq传递给apply的时候,不要加上圆括号
```

```python
sq
```

```python
#apply 传入 需要多个参数的函数
def my_exp(x,e):
    return x**e
```

```python
my_exp(2,3)
```

```python
ex = df['a'].apply(my_exp,e=2)
```

```python
ex
```

## apply 使用案例

```python
# 使用titanic数据集来介绍apply的用法
# 加载数据,使用info查看该数据集的基本特征
titanic = pd.read_csv('data/titanic.csv')
titanic.info()
```

```python
# 该数据集有891行,15列, 其中age 和 deck 两列中包含缺失值
# 可以使用apply计算数据中有多少null 或 NaN值
```

```python
# 缺失值数目
import numpy as np
def count_missing(vec):
    #根据值是否缺失获取一个由True/False组成的向量
    null_vec = pd.isnull(vec)
    # 得到null_vec中null值的个数
    # null值对应True, True为1
    null_count = np.sum(null_vec)
    #返回向量中缺失值的个数
    return null_count
```

```python
# 缺失值占比
def prop_missing(vec):
    # 计算缺失值的个数
    # 这里使用刚刚编写的count_missing函数
    num = count_missing(vec)
    #获得向量中元素的个数
    #也需要统计缺失值个数
    dem = vec.size
    return num/dem
```

```python
# 非缺失值占比
def prop_complete(vec):
    #先计算缺失值占的比例
    #然后用1减去缺失值的占比
    return 1-prop_missing(vec)
```

```python
# 把前面定义好的函数应用于数据的各列
titanic.apply(count_missing)
```

```python
titanic.apply(prop_missing)
```

```python
titanic.apply(prop_complete)
```

```python
#把前面定义好的函数应用于数据的各行
titanic.apply(count_missing ,axis = 1)
```

```python
titanic.apply(prop_missing ,axis = 1)
```

```python
titanic.apply(prop_complete ,axis = 1)
```

```python

titanic.apply(count_missing,axis = 1).value_counts()
```

## 向量化函数

```python
# 创建一个DataFrame
df = pd.DataFrame({'a':[10,20,30],'b':[20,30,40]})
```

```python
# 创建函数
def avg_2(x,y):
    return (x+y)/2
avg_2(df['a'],df['b'])

```

```python
# 修改函数
def avg_2_mod(x,y):
    if(x==20):
        return (np.NaN)
    else:
        return (x+y)/2
avg_2_mod(df['a'],df['b'])	
```

```python
# 使用np.vectorize将函数向量化
avg_2_mod_vec = np.vectorize(avg_2_mod)
avg_2_mod_vec(df['a'],df['b'])
```

```python
# 使用装饰器
@np.vectorize
def vec_avg_2_mod(x,y):
    if(x==20):
        return (np.NaN)
    else:
        return (x+y)/2
vec_avg_2_mod(df['a'],df['b'])
```
