# 语音识别案例
* 获取语音
* 调用AipSpeech.asr识别

```python
import pyaudio
from aip import AipSpeech
from pprint import pprint
```

```python
# 获取语音

RATE = 16000
CHUNK = 1024
DURATION = 10

# 创建PyAudio对象
pa = pyaudio.PyAudio()

# 创建stream 对象
stream = pa.open(format=pyaudio.paInt16,
                channels=1,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)
# 获取声音
frames = []
input('开始录音:')
for i in range(DURATION*RATE//CHUNK + 1):
    data = stream.read(CHUNK)
    frames.append(data)
print('录音结束')

# 拼接为完整的音频流
audio_buf = b''.join(frames)

# 关闭拼音流
stream.stop_stream()
stream.close()

pa.terminate()
```

```python
# 实例化AipSpeech 
aip = AipSpeech('11509036', 'gZtV8bLV68w8x5s1uURLRWyC', 'fBnTukj5a3jCrMTGdQzMznoIfE0IiPS4')

# 调用asr函数识别
result = aip.asr(audio_buf, 'wav', 16000, {'dev_pid':1537})
pprint(result)
```

```python
# 语音合成
text = '在阿根廷3:4法国的比赛中，姆巴佩攻入两球，并制造了法国首打破僵局的点球，他那场犹如天神下凡的表现，惊艳了整个足坛。'

options = {}
options['spd'] = 7
options['pit'] = 2
options['vol'] = 10
options['per'] = 3

result = aip.synthesis(text, options=options)

if not isinstance(result, dict):
    with open('audio.mp3', 'wb') as f:
        f.write(result)
```
